```python

interval_seconds = 1800 

start_time = datetime.datetime.now()
end_time = start_time + datetime.timedelta(hours=24)
current_index = 15  
predictions = []

while datetime.datetime.now() <= end_time:
    current_time = datetime.datetime.now()
    
    latest_data = fetch_latest_data(current_index)
    
    if latest_data.shape[0] < 15:
        print(f"Not enough data to make prediction at index {current_index}.")
    else:
      
        predicted_value = predict_with_lstm(model, latest_data)
        
      
        predicted_value = scaler.inverse_transform([[predicted_value]])[0][0]
        
       
        actual_value = scaler.inverse_transform([[latest_data[-1]]])[0][0]
        
     
        print(f"Time: {current_time}, Actual Value: {actual_value}, Predicted Value: {predicted_value}")
        predictions.append((current_time, actual_value, predicted_value))
        
        
        message = f"Time: {current_time}\nActual Value: {actual_value}\nPredicted Value: {predicted_value}"
        send_telegram_message(bot_token, group_chat_id, message)
    
   
    current_index += 1
    
    # Sleep until the next prediction interval
    next_prediction_time = current_time + datetime.timedelta(seconds=interval_seconds)
    sleep_duration = (next_prediction_time - datetime.datetime.now()).total_seconds()
    
    if sleep_duration > 0:
        time.sleep(sleep_duration)
```
